Обучите модель линейной регрессии на данных согласно вашему варианту. Выведите
угловой коэффициент, точку пересечения с осью ординат и точность.

Вариант 5: y = 6.8*x + random(7)

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def func(x):
    return 6.8 * x + np.random.random(x.shape) * 7


X = np.arange(-10, 11, 1).reshape(-1, 1)
y = func(X)

plt.scatter(X, y)
plt.show()
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(X, y)
y_pred = lr.predict(X)

print(f"Угловой коэф: {lr.coef_[0]}")
print(f"Пересечение: {lr.intercept_}")
print(f"Точность: {lr.score(X, y)}")

plt.plot(X, y_pred)
plt.scatter(X, y)
plt.show()
```
